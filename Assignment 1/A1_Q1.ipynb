#Data Preparation

As stated earlier, the more relevant features will be used for modeling and others will be dropped off

```python
# Retaining column with higher correlation values for red dataset
columns_used_red = ['volatile acidity','fixed acidity','chlorides','alcohol','sulphates','total sulfur dioxide']
X_red = data_red[columns_used_red]
y_red = data_red['quality']
```

```python
# Retaining column with higher correlation values for white dataset
columns_used_white = ['volatile acidity','fixed acidity','citric acid','residual sugar','chlorides','pH','alcohol','sulphates','free sulfur dioxide']
X_white = data_white[columns_used_white]
y_white = data_white['quality']
```

```python
#Splitting the red wine quality dataset
from sklearn.model_selection import train_test_split

X_train_red, X_test_red, y_train_red, y_test_red = train_test_split(X_red, y_red, test_size = 0.25, random_state = 42)

# determining the shapes of training and testing sets
print(X_train_red.shape)
print(y_train_red.shape)
print(X_test_red.shape)
print(y_test_red.shape)
```

```python
#Splitting the red wine quality dataset
X_train_white, X_test_white, y_train_white, y_test_white = train_test_split(X_white, y_white, test_size = 0.25, random_state = 42)

# determining the shapes of training and testing sets
print(X_train_white.shape)
print(y_train_white.shape)
print(X_test_white.shape)
print(y_test_white.shape)
```

Scaling the data with standard scalar for red dataset for centering it and with robust scalar for white dataset due to presence of outliers.

```python
#scaling the data of red wine quality dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_red = scaler.fit_transform(X_train_red)
X_test_red = scaler.transform(X_test_red)
```

```python
#Scaling the data of white wine quality dataset
from sklearn.preprocessing import RobustScaler
transformer = RobustScaler()
X_train_white = transformer.fit_transform(X_train_white)
X_test_white = transformer.transform(X_test_white)
```

#Modeling



 ### Lasso Regression



```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

lsreg = Lasso()
parameter = {'alpha':[0.001, 0.01,0.02,0.03,0.04,0.06,0.08, 0.1, 1]} # defining hyperparameters
lsreg_red_ = GridSearchCV(lsreg, parameter, scoring = 'neg_mean_squared_error')
lsreg_red_.fit(X_train_red, y_train_red)
print('Best hyperparameters are ', lsreg_red_.best_params_)
print('the poly coefficients are',lsreg_red_.best_estimator_.coef_)

#Testing of the model
lasso_final_red=lsreg_red_.best_estimator_
y_predict_red=lasso_final_red.predict(X_test_red)

#Error prediction
print("For Red wine quality dataset:")
print('MAE:', mean_absolute_error(y_test_red, y_predict_red))
print('MSE:', mean_squared_error(y_test_red, y_predict_red))
print('RMSE:',np.sqrt(mean_squared_error(y_test_red, y_predict_red)))
print('r^2 Score:', r2_score(y_test_red,y_predict_red))
```

```python
parameter = {'alpha':[0.001, 0.01,0.02,0.03,0.04,0.06,0.08, 0.1, 1]} # defining hyperparameters
lsreg_white_ = GridSearchCV(lsreg, parameter, scoring = 'neg_mean_squared_error')
lsreg_white_.fit(X_train_white, y_train_white)
print('Best hyperparameters are ', lsreg_white_.best_params_)
print('the poly coefficients are',lsreg_white_.best_estimator_.coef_)

#Testing of the model
lasso_final_white=lsreg_white_.best_estimator_
y_predict_white=lasso_final_white.predict(X_test_white)

#Error prediction
print("For Red wine quality dataset:")
print('MAE:', mean_absolute_error(y_test_white, y_predict_white))
print('MSE:', mean_squared_error(y_test_white, y_predict_white))
print('RMSE:',np.sqrt(mean_squared_error(y_test_white, y_predict_white)))
print('r^2 Score:', r2_score(y_test_white,y_predict_white))
```

Minima is achieved at lower values of $\alpha$ in both the cases.

The parameter $\alpha$ is used for feature reduction, estimating the sparse coefficients. $\alpha$ = 0 is similar to linear regression

### Elastic Net

```python
# import model 
from sklearn.linear_model import ElasticNet 
#Red dataset  
# Train the model 
e_net = ElasticNet(alpha = 0.2,l1_ratio = 0.25) 
e_net.fit(X_train_red, y_train_red) 
# calculate the prediction and mean square error 
y_pred_elastic_red = e_net.predict(X_test_red) 
print("For Red wine quality dataset:")
print('MAE:', mean_absolute_error(y_test_red, y_pred_elastic_red))
print('MSE:', mean_squared_error(y_test_red, y_pred_elastic_red))
print('RMSE:',np.sqrt(mean_squared_error(y_test_red, y_pred_elastic_red)))
print('r^2 Score:', r2_score(y_test_red,y_pred_elastic_red))

#White dataset
# Train the model 
e_net = ElasticNet(alpha = 0.2,l1_ratio = 0.25) 
e_net.fit(X_train_white, y_train_white) 
# calculate the prediction and mean square error 
y_pred_elastic_white = e_net.predict(X_test_white) 
print("\nFor White wine quality dataset:")
print('MAE:', mean_absolute_error(y_test_white, y_pred_elastic_white))
print('MSE:', mean_squared_error(y_test_white, y_pred_elastic_white))
print('RMSE:',np.sqrt(mean_squared_error(y_test_white, y_pred_elastic_white)))
print('r^2 Score:', r2_score(y_test_white,y_pred_elastic_white))
```

The two parameter $\alpha$ and l1 ratio is used to set the coefficients of l1 norm and l2 norm in regression problem.

The elastic net is a combination of lasso and ridge regression. The parameter $\alpha$ in both the models are same. Setting the l1 ratio in elastic net model to 1, the elastic net model is treated as lasso regression model.

### Random Forest

```python
# Fitting Random Forest Regression to the red dataset
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 350, random_state = 0)
regressor.fit(X_train_red,y_train_red)
#Predicting the result
y_pred_rf_red = regressor.predict(X_test_red)
# calculating the training and testing accuracies
print("For Red wine quality dataset:")
print("Training accuracy :", regressor.score(X_train_red, y_train_red))
print("Testing accuracy :", regressor.score(X_test_red, y_test_red))
print('MAE:', mean_absolute_error(y_test_red, y_pred_rf_red))
print('MSE:', mean_squared_error(y_test_red, y_pred_rf_red))
print('RMSE:',np.sqrt(mean_squared_error(y_test_red, y_pred_rf_red)))
print('r^2 Score:', r2_score(y_test_red,y_pred_rf_red))

# Fitting Random Forest Regression to the white dataset
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 350, random_state = 0)
regressor.fit(X_train_white,y_train_white)
#Predicting the result
y_pred_rf_white = regressor.predict(X_test_white)
# calculating the training and testing accuracies
print("\nFor White wine quality dataset:")
print("Training accuracy :", regressor.score(X_train_white, y_train_white))
print("Testing accuracy :", regressor.score(X_test_white, y_test_white))
print('MAE:', mean_absolute_error(y_test_white, y_pred_rf_white))
print('MSE:', mean_squared_error(y_test_white, y_pred_rf_white))
print('RMSE:',np.sqrt(mean_squared_error(y_test_white, y_pred_rf_white)))
print('r^2 Score:', r2_score(y_test_white,y_pred_rf_white))
```

The parameter n estimator is used to decide the number of trees in the regressor to be used for modeling the data.

### Support Vector Regression

```python
from sklearn.svm import SVR
svr_red = SVR(kernel = 'rbf',C = 0.7,epsilon = 0.5)
svr_red.fit(X_train_red,y_train_red)

# Predicting a new result

y_pred_svr_red = svr_red.predict(X_test_red)

#Calculating errors
print("\nFor Red wine quality dataset:")
print('MAE:', mean_absolute_error(y_test_red, y_pred_svr_red))
print('MSE:', mean_squared_error(y_test_red, y_pred_svr_red))
print('RMSE:',np.sqrt(mean_squared_error(y_test_red, y_pred_svr_red)))
print('r^2 Score:', r2_score(y_test_red,y_pred_svr_red))

#For white wine quality dataset
svr_white = SVR(kernel = 'rbf',C = 0.8,epsilon = 0.7)
svr_white.fit(X_train_white,y_train_white)

# Predicting a new result

y_pred_svr_white = svr_white.predict(X_test_white)

#Calculating errors
print("\nFor White wine quality dataset:\n")
print('MAE:', mean_absolute_error(y_test_white, y_pred_svr_white))
print('MSE:', mean_squared_error(y_test_white, y_pred_svr_white))
print('RMSE:',np.sqrt(mean_squared_error(y_test_white, y_pred_svr_white)))
print('r^2 Score:', r2_score(y_test_white,y_pred_svr_white))
```

In SVR, the 'c' parameter trades off correct classification of training examples against maximization of the decision functions margin.
